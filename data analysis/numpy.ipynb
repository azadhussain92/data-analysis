# Element by operation

```python
x=np.array([2,4,6,8])
y=np.array([5,4,6,3])
```

```python
print(x,y,sep='\n')
```

```python
print(len(x),len(y),sep= '\n')
```

```python
# arithematic poperator
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x//y)
print(x%y)
```

```python
# relational operator

print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
```

```python
#bitwise operators

print(x&y)
print(x|y)
print(x^y)
print(x<<y)
print(x>>y)
print(~x)

```

```python
#logical operators
# x==y 
(x==y).all() and (x>y).any()
# np.logical_and(x==y,x>y)
```

```python
np.logical_and(x==y,x>y)
```

```python
np.logical_or(x==y,x>y)
```

```python
np.logical_not(x==y)
```

```python
np.logical_xor(x==y,x>y)
```

```python
# membership operator
8 in x
```
