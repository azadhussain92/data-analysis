### rand()

```python
rd.rand()
```

```python
rd.rand(3)
```

```python
rd.rand(3,2)
```

### randn()
return a simple (or sample ) from the "standard normal"distribution.
mean nearly eqal to std nearly equal to 1

```python
rd.randn() # if you dont pass the any value its fetch the 1 value
```

```python
rd.randn(3,)
```

```python
x=rd.randn(3,4)
x
```

```python
print(x.mean())
print(x.std())
```

### seed(seed_val)
