```python
import pandas as pd
import numpy.random as rd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rd.seed(35)
g=['male','female']
rd.choice(g,size=40)
```

```python
rd.seed(25)
g=['male','female']
gender=rd.choice(g,size=40)
```

```python
rd.seed(25)
rd.randint(15,25,size=40)
```

```python
rd.seed(25)
age=rd.randint(20,71,size=40)
```

```python
rd.seed(25)
location=rd.choice(['Morocco','Turkey','Saudi Arabia'],size=40)
```

```python
d={'GENDER':gender,'AGE':age,'LOCATION':location}
```

```python
df=pd.DataFrame(d)
df
```

```python
plt.Figure(figsize=(5,3))
sns.countplot(df,x='GENDER')
plt.show()
```

```python
plt.Figure(figsize=(5,3))
c=sns.countplot(df,x='GENDER',color='yellow')
plt.title('no of male and female')
c.bar_label(c.containers[0],padding=-12)
plt.show()
```
