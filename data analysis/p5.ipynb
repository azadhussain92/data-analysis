# copy() and 

```python
x1=np.array([1,2,3])
x2=np.copy(x1)
print(x1,x2,sep='\n')
print("after modification of x1")
x1[0]=100
print(x1,x2,sep='\n')
print("x1 and x2 are independent variable")
```

```python
y1=np.array([1,2,3])
y2=np.copy(y1)
print(x1,x2,sep='\n')
print("after modification of y1")
y1[0]=100
print(y1,y2,sep='\n')
print("y1 and y2 are dependent variable")
```

# np.reshape()

```python
z1=np.arange(1,13,1)
print(f'z1={z1} and length of array is  {z1.size}')
# choose the factores of 12(len) for reshaping
# (1,12) (12,1) (2,6) (6,2) (3,4) (4,3)
# (1,12,1) (12,1,1) (2,3,2) (2,2,3)
# (1,12,1,1) (12,1,1,1) (2,3,2,1) (2,2,3,1)

print(np.reshape(z1,shape=((2,6))),end='\n\n')
print(np.reshape(z1,shape=((2,3,2))),end='\n\n')
```

```python
z2=np.array([[10,20,30],[40,50,60]])  # (2,3)
print(z2,end='\n\n')
res=np.reshape(z2,shape=-1) # ndarray to 1d array
print(res)
# print(f'\n {'-'*25} \n shallow copy or view ?\n {'-'*25} \n')
print("shallow copy or view ? ")
z2[0,0]=100
print(z2,res ,sep='\n\n')  # reshape and return view
```

```python
np.reshape(z2,shape=(3,2))   # dim remain same but shape is changed
```

# np.ravel

```python
z3=np.array([[10,20,30],[40,50,60]])
res=np.ravel(z3)  # ndarray to 1darray (return view of z3)
z3[0,0]=500
print(z3,res,sep='\n\n')
```

# agg with 2darray()

```python
a1=np.array([[10,20],[30,40]])
print(np.sum(a1,axis=0))  # 0th element of each braket 1st element of each bracket
# [10+30,20+40]===>[40,60]
#------------------------------------------------------------------
print(np.sum(a1,axis=1))  # inner element are addad up ( withi the same bracket )
# [10+20 , 30+40]----->[30,70]
```
