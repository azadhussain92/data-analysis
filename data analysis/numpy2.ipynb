```python
import numpy as np
```

```python
# scaler array

a0= np.array(67) # 67 or 6.9 or 6+9j or 'programming' any permitive data type
print(a0)
print(a0.ndim)  # no_of diamensional (array diamensional)
print(a0.dtype) # data type of array elements
print(a0.itemsize) # memory utilized by single element(in n bytes)
print(a0.nbytes) #memory utilized by array (in bytes)
print(a0.shape) # no of element of each diamensional
print(a0.size) # total element
```

```python
# one  diamensional array

a1= np.array([10,20,30])
print(a1)
print(a1.ndim)
print(a1.size) # total element
print(a1.dtype) # data type of array elements
print(a1.itemsize) # memory utilized by single element(in n bytes)
print(a1.nbytes) #memory utilized by array (in bytes)
print(a1.shape) # no of element of each diamensional
```
