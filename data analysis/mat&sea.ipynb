```python
import seaborn as sns
import pandas as pd
import numpy as np
import numpy.random as rd 
import matplotlib.pyplot as plt
```

```python
rd.seed(55)

g=['Female','Male',]
rd.choice(g,size=50)   
```

```python
rd.seed(55)

g=['Female','Male']
gender=rd.choice(g,size=50)  #array of 50 random value
```

```python
rd.seed(31)
rd.randint(20,30,size=50)
```

```python
rd.seed(31)
age=rd.randint(20,81,size=50)# array of 50 random int b/w 20-80
```

```python
rd.seed(31)
location=rd.choice(['Bangaluru','Chennai','Delhi'],size=50)
```

```python
rd.seed(50)
expenses=rd.randint(500,1000,50)
```

```python
d={'GENDER':gender,'AGE':age,'LOCATION':location,'EXPENSES':expenses}
```

```python
df=pd.DataFrame(d)
df
```
