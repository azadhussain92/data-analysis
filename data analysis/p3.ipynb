```python
# to create age_category we can use either pd.cut() or df['age'].apply()
# step1: create user_define function()
# step:2 pass the user_define function to apply() as a argument

def age_category(a):
    if a >= 18 and a<=25:
        return 'Teen'
    elif a >= 26 and a <=59:
        return 'Adult'
    elif a >=60 and a <= 100:
        return 'Senior'
```

```python
df['age'].apply(age_category)

#apply() method taken the value from age_col and pass it to age_category function . it age value satisfied the condition then it return the els none(default)(teen, adult,senior)
```

```python
df['age_category']=df['age'].apply(age_category)
```
