```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Create Data

```python
data={
    'price':['3$','50rs','rs50','50rs+50']
}
```

## Create DataFrame

```python
df=pd.DataFrame(data)
print("Original Data:")
df
```

## Clean data

```python
df['clean_price'] = (
    df['price']
    .str.replace('rs','',regex=False)
    .str.replace('$','',regex=False)
    .str.replace('+','',regex=False)
)
```

## Convert to numeric

```python
df['clean_price']=pd.to_numeric(df['clean_price'],errors='coerce')
print('\nclean_data:')
print(df)
```

## Fill NaN values

```python
df['clean_price']=df['clean_price'].fillna(df['clean_price'].mean())
print(df)
```

```python
print(df.columns)
```

## Visualization

```python
plt.Figure(figsize=(5,3))
sns.barplot(x=df.index,y=df['clean_price'])
plt.title("clean price data")
plt.xlabel("index")
plt.ylabel("price")
```

```python
plt.Figure(figsize=(5,3))
sns.scatterplot(x=df.index,y=df['clean_price'])
plt.show()
```
