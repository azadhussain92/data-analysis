## Method to create array

1. np.arange()

2. np.linespace()

3. np.ones()

4. np.zeros()

5. np.full()

6. np.eye()

```python
# np.arange(): serese of eqally space numbers

np.arange(start=0,stop=10,step=2) 
 # start(inclusuve and default -->0)
 # stop(exclusive)
 # step(default-->1)
 # return type --->ndarray(int)
```

```python
# np.linspace()  : return of arary of qyally spaced number b/w start (inclusive) and stop (inclusive)
# num=50(default it generate the number)

np.linspace(start=1,stop=20,num=5)
# np.linspace(start=1,stop=20,num=10)
```

```python
np.linspace(start=1,stop=20,num=5,retstep=True)
```

```python
np.linspace(start=1,stop=20,num=5,endpoint=False,retstep=True)
```

```python
# np.ones() : return array of ones in float

np.ones(shape=(3,))
```

```python
np.ones(shape=(3,2))
```

```python
np.ones(shape=(3,2),dtype=int)
```

```python
# np.zeros()  : return array of zeros in float
np.zeros(shape=(3,))
```

```python
np.zeros(shape=(3,2))
```

```python
# np.zeros(shape=(3,),dtype=float)
```

```python
# np.full() : return the array of values 
# same value will be filled for the given shape

np.full(shape=(3,),fill_value=100)
np.full(shape=(3,2),fill_value=100)
```

```python
# np.eye(): return the identity materix

np.eye(N=3)  # n-->row ,m-->col(default m=n)
# sqared materix 
```
